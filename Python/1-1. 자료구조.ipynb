# 변수, 상수, 계수란?

image.png

- 변수(variable)
  - 변수의 변은 변하다라는 뜻을 가지고 있습니다. 
  - 즉, 프로그래밍 중 데이터가 변하는 경우에는 변수를 사용합니다.
- 상수(contant)
  - 상수는 항상 같은 수, 변하지 않는 수를 말합니다.
  - 즉, 프로그래밍 중 데이터가 어떤 경우에도 변하지 않는 경우에 상수를 사용합니다.
- 계수(coefficient)
  - coefficient에서 co는 함께라는 의미이고, efficient는 효과적인/효율적인이란 뜻입니다. 
  - 즉, 상수 중에서 변수와 관련이 있는 것을 계수라고 합니다.
  - 머신러닝/딥러닝에서 많이 사용됩니다.

## enum
일반적으로 서로 관련이 있는 여러 개의 상수의 집합을 정의할 때 사용된다.

```python
import enum
```

```python
class Rainbow(enum.Enum):
  Red = (enum.auto(), 0)
  Orange = (enum.auto(), 1)
  Yellow = (enum.auto(), 2)
  Green = (enum.auto(), 3)
  Blue = (enum.auto(), 4)
  Navy = (enum.auto(), 5)
  Purple = (enum.auto(), 6)
```

```python
Rainbow['Red']
```

```python
Rainbow.Red
```

```python
Rainbow.Red.name
```

```python
Rainbow.Red.value[0]
```

```python
Rainbow.Red.value[1]
```

```python
Rainbow.Red == Rainbow.Red
```

```python
Rainbow.Red.name == 'Red'
```

```python
Rainbow.Red is Rainbow.Green
```

```python
Rainbow.Red == Rainbow.Green
```

```python
Rainbow.__len__
```

```python
for e in Rainbow:
  print((e.name, e.value[1]))
```

- Direct: 직접 할당

## 인덱싱
인덱싱(indexing)이란 무엇인가를 가리킨다는 의미이다.   
이때 주의할 점은 파이썬은 숫자를 셀때, 0부터 시작한다는 점이다.

```python
a = "Hello World!"
a
```

Hello World!라는 문자에서   
H는 0인덱스, e는 1인덱스, l은 2인덱스, 1은 3인덱스, o는 4인덱스,   
빈칸은 5인덱스, W은 6인덱스, o은 7인덱스, r은 8인덱스,   
l은 9인덱스, d는 10인덱스, !은 11인덱스를 뜻한다.   

파이썬에서 인덱싱을 표현하는 방법은 `변수명[인덱스]`이다.

```python
a[0]
```

```python
a[11]
```

`-1`의 인덱싱 의미는 뒤에서 부터 1번째 값을 의미한다.   
즉, `-숫자`의 인덱싱 의미는 뒤에서 부터 숫자번째 값을 의미한다.

```python
a[-1]
```

```python
a[-4]
```

## 슬라이싱
슬라이싱(slicing)은 무엇인가를 잘라낸다는 의미이다.  

```python
a = "Hello world!"
a
```

```python
a[0] + a[1] + a[2] # H + e + l
```

위의 코드처럼 인덱싱과 문자열 더하기를 통해서 표현할 수도 있지만,   
아래와 같이 간단하게 슬라이싱으로 표현할 수도 있다.   

슬라이싱 사용법은 `변수명[시작 인덱싱 번호: 끝 인덱싱 번호]`로 표현할 수 있으며, 이때, 시작 인덱싱은 포험되지만, 끝 인덱싱은 포함되지 않는다.

```python
a[0:3] # a[0] + a[1] + a[2]
```

```python
a[:3] # 시작 인덱싱 번호가 0이면 생략 가능
```

```python
a[6:12] # a[6] + a[7] + a[8] + a[9] + a[10] + a[11]
```

```python
a[6:] # 끝 인덱싱 번호를 생략하면 문장 끝까지 표현
```

```python
a[:-6] # 음수 표현
```

`문자열[시작:끝:간격]` -> 간격은 생략 가능

```python
a
```

```python
a[0:-1:2]
```

```python
a = "현재 온도는 18도 입니다."
a
```

```python
a[:7]
```

```python
a[9:]
```

```python
a[:7]+"20"+a[9:] # 슬라이싱 + 문자열 더하기
```

## 문자열 포매팅

문자열 안에서 숫자를 넣고 싶은 자리에 `%d`를 사용한다.

```python
print("I eat %d apples." % 3)
```

문자열 안에서 문자를 넣고 싶은 자리에 `%s`를 사용한다.

```python
print("I eat %s apples." % "five")
```

```python
number = 10
day = "three"

print("I ate %d apples. so I was sick for %s days." % (number, day))
```

소수점 표현할 때는 `%0.숫자f`이다.    
즉, 소수점 3째자리까지 표현하려면 `%0.3f`으로 사용한다.

```python
print("파이는 %0.3f" % 3.1428)
```

### Format

```python
print("I eat {} apples".format(3))
```

```python
number = 10
day = "three"

print("I ate {0} apples. so I was sick for {1} days.".format(number, day))
```

```python
number = 10
day = "three"

print("I ate {number} apples. so I was sick for {day} days.".format(number=number, day=day))
```

### f 문자열 포메딩

```python
number = 10
day = "three"

print(f"I ate {number} apples. so I was sick for {day} days.")
```
